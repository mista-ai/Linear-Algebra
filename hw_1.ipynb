__2.__ Найдите нормы векторов $(4, 2, 4)$ и $(12, 3, 4)$ и угол между ними.


```python
from numpy.linalg import norm
a = np.array([4, 2, 4])
b = np.array([12, 3, 4])

print(f'Норма вектора (4, 2, 4): {norm(a)}')
print(f'Норма вектора (12, 3, 4): {norm(b)}')
print(f'Угол между векторами: {np.arccos(np.dot(a, b)/norm(a)/norm(b)) * 180}')
```

__3.__ Определите, будет ли линейное пространство евклидовым, если за скалярное произведение принять:<br>

Линейное пространство над полем вещественных чисел называется _евклидовым пространством_, если на нём введено правило, согласно которому каждой упорядоченной паре векторов $x$ и $y$ поставлено в соответствие вещественное число, называемое _скалярным произведением_ и обозначаемое символом $(x, y)$.

Указанное правило должно подчиняться четырём аксиомам:<br>
1) $(x,y)=(y,x).$<br>
2) $(\lambda x, y)=\lambda(x, y).$<br>
3) $(x_{1}+x_{2},y)=(x_{1},y)+(x_{2},y).$<br>
4) $(x, x)\geq0$, причём $(x,x)=0\Leftrightarrow x=0.$

__а)__ произведение длин векторов;<br>
1) $(x,y)=||x||\cdot||y||=||y||\cdot||x||=(y,x).$<br>
2) $(\lambda x, y)=||\lambda x||\cdot||y||=|\lambda| ||x||\cdot||y||=|\lambda|(y, x)\neq\lambda(y, x).$<br>

Т.к. условие 2-й аксиомы не выполняется, такое пространство не явл-ся евклидовым<br>
__Ответ: не является__

__б)__ утроенное обычное скалярное произведение векторов?<br>
1) $(x,y)=3(x, y)=3(y, x)=(y,x).$<br>
2) $(\lambda x, y)=3(\lambda x, y)=3\lambda(x, y)=\lambda(x, y).$<br>
3) $(x_{1}+x_{2},y)=3(x_1+x_2, y)=3((x_1, y)+(x_2, y))=3(x_1, y)+3(x_2,y)=(x_{1},y)+(x_{2},y).$<br>
4) $(x, x)=3(x, x)\geq0$, причём $3(x,x)=0\Leftrightarrow x=0.$<br>

Все аксиомы выполняются, поэтому пространство евклидово<br>
__Ответ: является__

$$(e_1, e_2)=1/\sqrt{2}*1/\sqrt{2} + (-1/\sqrt{2})*1/\sqrt{2}+0*0=1/2-1/2+0=0$$
$$(e_1, e_3)=1/\sqrt{2}*0 + (-1/\sqrt{2})*0+0*1=0+0+0=0$$
$$(e_2, e_3)=1/\sqrt{2}*0 + 1/\sqrt{2}*0+0*1=0+0+0=0$$
Проверим, что векторы имеют единичную длину
$$(e_1, e_1) = 1/\sqrt{2}*1/\sqrt{2} + (-1/\sqrt{2})*(-1/\sqrt{2})+0*0=1/2+1/2+0=1$$
$$(e_2, e_2)=1/\sqrt{2}*1/\sqrt{2} + 1/\sqrt{2}*1/\sqrt{2}+0*0=1/2+1/2+0=1$$
$$(e_3, e_3)=0*0 + 0*0+1*1=0+0+1=1$$
Все условия определения ортонормированного базиса выполняются.<br>
__Ответ: образуют__
